```python
from bs4 import BeautifulSoup
```

```python
html_path ='''<h1>Welcome to Our Company</h1>
<h2>Web Site Main Ingredients:</h2>
<p>Pages(HTML)</p>
<p>Style Sheets(CSS)
<b>SUPRISE</b>
</p>
<p>Computer Code(JavaScript)</p>
<p>Live Data(Files and Databases)</p>'''
```

```python
doc = BeautifulSoup(html_path ,"html.parser")
```

```python
print(doc)
```

```python
print(doc.prettify())
```

# To Find a Tag in The HTML

```python
tag =doc.h1
print(tag)
```

# Lets access the string inside this tag

```python
print(tag.string)
```

# Let change this string

```python
tag.string="Hello Good Evening"
print(tag)
```

```python
print(doc.prettify())
```

# Now lets find all tags which occurs in tag or pages tag

```python
tags = doc.find("p")
```

```python
print(tags)
```

```python
tags2 = doc.find_all("p")
print(tags2)
```
